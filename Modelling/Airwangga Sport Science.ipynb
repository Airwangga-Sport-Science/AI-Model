```python
def create_statistic_bar_chart(data, selected_columns, statistics):
    plt.figure(figsize=(14, 10))

    filtered_data = data[selected_columns]

    statistical_data = filtered_data.describe().transpose()
    statistical_data = statistical_data[statistics]
    statistical_data = statistical_data.reset_index()

    statistical_data = pd.melt(statistical_data, id_vars="index", var_name="Statistic", value_name="Value")
    statistical_data = statistical_data.rename(columns={"index": "Column"})

    ax = sns.barplot(x="Statistic", y="Value", hue="Column", data=statistical_data, palette="Set3")
    plt.title("Statistical Description")
    plt.xlabel("Statistic")
    plt.ylabel("Value")
    plt.xticks(rotation=0)
    plt.legend(title="Column")

    for p in ax.patches:
        value = p.get_height()
        x_position = p.get_x() + p.get_width() / 2.
        y_position = value + 1  # Adjust the vertical position
        ax.annotate(f'{value:.2f}', (x_position, y_position),
                    ha='center', va='center', fontsize=10, color='black')

    plt.show()
```

```python
def plot_player_value_distribution(df, positions, thresholds, legend_fontsize):
    """
    Plot the distribution of player values for each position in a grid of separate histograms,
    add red lines at the specified thresholds, and display the number of players above each threshold for each position.

    Parameters:
    df (DataFrame): The DataFrame containing player positions and values.
    positions (list): A list of position column names.
    thresholds (list): A list of threshold values.
    legend_fontsize (int): Font size for the legend text.

    Returns:
    None
    """
    num_positions = len(positions)
    num_cols = 3  # Number of columns in the grid
    num_rows = (num_positions + num_cols - 1) // num_cols  # Calculate the number of rows needed

    # Define colors for the thresholds
    threshold_colors = ['orange', 'red']

    # Create a grid of subplots
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(25, 20))
    fig.suptitle('Distribution of Player Values by Position', fontsize=16)

    for i, position in enumerate(positions):
        row = i // num_cols
        col = i % num_cols

        position_data = df[position]

        ax = axes[row, col]
        ax.hist(position_data, bins=10, alpha=0.5)
        
        # Add red lines at the specified thresholds and count players above each threshold
        for idx, threshold in enumerate(thresholds):
            above_threshold_count = (position_data > threshold).sum()
            ax.axvline(x=threshold, color=threshold_colors[idx], linestyle='--', label=f'{threshold} (Above: {above_threshold_count})')
        
        ax.set_title(f'{position}')
        ax.set_xlabel('Player Value')
        ax.set_ylabel('Number of Players')
        
        # Set the legend font size
        legend = ax.legend()
        for item in legend.get_texts():
            item.set_fontsize(legend_fontsize)

    # Hide any empty subplots if the number of positions is not a multiple of num_cols
    for i in range(len(positions), num_rows * num_cols):
        row = i // num_cols
        col = i % num_cols
        fig.delaxes(axes[row, col])

    plt.tight_layout()
    plt.subplots_adjust(top=0.9)
    plt.show()
```

```python
def plot_position_percentage_above_thresholds(df, thresholds, positions_to_visualize):
    """
    Plot the percentage of players in selected soccer positions with values above the specified thresholds compared to
    the total number of players.

    Parameters:
    df (DataFrame): The DataFrame containing player positions and values.
    thresholds (list): A list of threshold values.
    positions_to_visualize (list): A list of positions to visualize. If None, all positions will be visualized.

    Returns:
    None
    """
    if positions_to_visualize is None:
        positions_to_visualize = df.columns
    
    position_percentages = {position: [] for position in positions_to_visualize}

    for threshold in thresholds:
        for position in positions_to_visualize:
            above_threshold_count = (df[position] > threshold).sum()
            total_count = len(df)
            percentage = (above_threshold_count / total_count) * 100
            position_percentages[position].append(percentage)

    num_positions = len(positions_to_visualize)
    width = 0.35
    x = range(num_positions)

    fig, ax = plt.subplots(figsize=(12, 6))
    for i, threshold in enumerate(thresholds):
        bars = ax.bar(
            [pos + width * i for pos in x],
            [position_percentages[position][i] for position in positions_to_visualize],
            width,
            label=f'Above {threshold}',
        )
        
        # Add percentage labels above the bars
        for bar in bars:
            height = bar.get_height()
            ax.annotate(f'{height:.2f}%', 
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(8, 3),  # 8, 3 points offset
                        textcoords="offset points",
                        ha='center', va='bottom')

    ax.set_xlabel('Soccer Positions')
    ax.set_ylabel('Percentage of Players')
    ax.set_title('Percentage of Players in Selected Soccer Positions Above Thresholds')
    ax.set_xticks([pos + width for pos in x])
    ax.set_xticklabels(positions_to_visualize)
    ax.legend(title='Thresholds')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
def plot_player_skill_distribution(df, skills, legend_fontsize):
    """
    Plot the distribution of player skills for each skill in a grid of separate histograms,
    add red lines at the above average and above 75th percentile thresholds,
    and display the number of players above each threshold for each skill.

    Parameters:
    df (DataFrame): The DataFrame containing player skills and values.
    skills (list): A list of skill column names.
    legend_fontsize (int): Font size for the legend text.

    Returns:
    None
    """
    num_skills = len(skills)
    num_cols = 3  # Number of columns in the grid
    num_rows = (num_skills + num_cols - 1) // num_cols  # Calculate the number of rows needed

    # Define colors for the thresholds
    threshold_colors = ['orange', 'red']

    # Create a grid of subplots
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(25, 20))
    fig.suptitle('Distribution of Player Skills by Skill', fontsize=16)

    for i, skill in enumerate(skills):
        row = i // num_cols
        col = i % num_cols

        skill_data = df[skill]

        ax = axes[row, col]
        ax.hist(skill_data, bins=10, alpha=0.5)
        
        # Calculate the above average and above 75th percentile thresholds
        above_60th_percentile_threshold = skill_data.quantile(0.60)
        above_90th_percentile_threshold = skill_data.quantile(0.90)
        
        # Add red lines at the above 60th percentile and above 90th percentile thresholds
        for idx, threshold in enumerate([above_60th_percentile_threshold, above_90th_percentile_threshold]):
            above_threshold_count = (skill_data > threshold).sum()
            ax.axvline(x=threshold, color=threshold_colors[idx], linestyle='--', label=f'Above {threshold:.2f} (Above: {above_threshold_count})')
        
        ax.set_title(f'{skill}')
        ax.set_xlabel('Skill Value')
        ax.set_ylabel('Number of Players')
        
        # Set the legend font size
        legend = ax.legend()
        for item in legend.get_texts():
            item.set_fontsize(legend_fontsize)

    # Hide any empty subplots if the number of skills is not a multiple of num_cols
    for i in range(len(skills), num_rows * num_cols):
        row = i // num_cols
        col = i % num_cols
        fig.delaxes(axes[row, col])

    plt.tight_layout()
    plt.subplots_adjust(top=0.9)
    plt.show()
```

```python
def visualize_top_positive_correlations(df, positions_to_visualize, skills, top_n):
    """
    Visualize the top N highest positive correlations between soccer player position value and skills for each position.

    Parameters:
    df (DataFrame): The DataFrame containing player positions and skills.
    positions_to_visualize (list): A list of positions to visualize.
    skills (list): A list of skills to consider.
    top_n (int): Number of top positive correlations to visualize (default is 5).

    Returns:
    None
    """
    num_positions = len(positions_to_visualize)
    num_cols = 3  # Number of columns in the grid
    num_rows = (num_positions + num_cols - 1) // num_cols  # Calculate the number of rows needed

    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10 * num_rows))
    fig.suptitle(f'Top {top_n} Highest Positive Correlations between Player Position Value and Skills', fontsize=16)

    for i, position in enumerate(positions_to_visualize):
        row = i // num_cols
        col = i % num_cols

        # Calculate the correlations between position value and each skill
        correlations = df[skills].corrwith(df[position])

        # Sort correlations in descending order
        top_correlations = correlations.sort_values(ascending=False).head(top_n)

        ax = axes[row, col]
        ax.barh(top_correlations.index, top_correlations.values, color='dodgerblue')
        ax.set_xlabel('Correlation Value')
        ax.set_title(f'Top {top_n} Positive Correlations for {position}')
        ax.invert_yaxis()  # Invert the y-axis for better visualization

    # Hide any empty subplots if the number of positions is not a multiple of num_cols
    for i in range(len(positions_to_visualize), num_rows * num_cols):
        row = i // num_cols
        col = i % num_cols
        fig.delaxes(axes[row, col])

    plt.tight_layout()
    plt.subplots_adjust(top=0.9)
    plt.show()
```

```python
positions_to_visualize = ['GK', 'DL', 'DC', 'DR', 'WBL', 'WBR', 'DM', 'MR', 'ML', 'MC', 'AML', 'AMC', 'AMR', 'ST']

skills_to_visualize = ["Corners", "Crossing", "Dribbling", "Finishing", "First Touch",
    "Free Kick Taking", "Heading", "Long Shots", "Passing", "Tackling",
    "Technique", "Concentration", "Vision", "Decision", "Determination",
    "Position.1","Teamwork", "Balance", "Natural Fitness", "Pace",
    "Stamina", "Strength", "Communication", "Reflexes", "Left Foot", "Right Foot"]

top_n = 10

visualize_top_positive_correlations(df, positions_to_visualize, skills_to_visualize, top_n)
```

```python
def top_skills_by_position(df, positions, skills, top_n=10):
    """
    Create a dictionary that associates each soccer player position with its top N skills based on correlations.

    Parameters:
    df (DataFrame): The DataFrame containing player positions and skills.
    positions (list): A list of positions to consider.
    skills (list): A list of skills to consider.
    top_n (int): Number of top skills to associate with each position (default is 10).

    Returns:
    dict: A dictionary where keys are positions, and values are lists of top N skills.
    """
    position_skills_dict = {}

    for position in positions:
        correlations = df[skills].corrwith(df[position])
        top_skills = correlations.sort_values(ascending=False).head(top_n).index.tolist()
        position_skills_dict[position] = top_skills

    return position_skills_dict

# Usage example:
positions_to_visualize = ['DL', 'DC', 'DR', 'WBL', 'WBR', 'DM', 'MR', 'ML', 'MC', 'AML', 'AMC', 'AMR', 'ST']

skills = ["Corners", "Crossing", "Dribbling", "Finishing", "First Touch",
    "Free Kick Taking", "Heading", "Long Shots", "Passing", "Tackling",
    "Technique", "Concentration", "Vision", "Decision", "Determination",
    "Position.1","Teamwork", "Balance", "Natural Fitness", "Pace",
    "Stamina", "Strength", "Communication", "Reflexes", "Left Foot", "Right Foot"]

top_skills = top_skills_by_position(df, positions_to_visualize, skills, top_n=10)
```

```python
import numpy as np
from collections import Counter
```

```python
def calculate_position_percentages(skill_positions, position_categories):
    skill_percentage_dict = {}

    for skill, positions in skill_positions.items():
        category_counts = Counter()
        
        for category, category_positions in position_categories.items():
            category_count = sum(1 for pos in positions if pos in category_positions)
            category_percentage = (category_count / len(category_positions)) * 100
            category_counts[category] = category_percentage

        skill_percentage_dict[skill] = category_counts

    return skill_percentage_dict
```

```python
def plot_position_percentages(skill_percentage_dict):
    skills = list(skill_percentage_dict.keys())
    num_skills = len(skills)
    num_columns = 3  # Number of columns in the grid
    num_rows = (num_skills + num_columns - 1) // num_columns

    # Define colors for each skill
    colors = plt.cm.viridis(np.linspace(0, 1, num_skills))

    fig, axes = plt.subplots(num_rows, num_columns, figsize=(15, 6 * num_rows))
    fig.suptitle('Position Percentage for Each Skill', fontsize=16)

    for i, skill in enumerate(skills):
        row = i // num_columns
        col = i % num_columns
        percentages = skill_percentage_dict[skill]
        categories = list(percentages.keys())
        values = list(percentages.values())

        ax = axes[row, col]
        ax.barh(categories, values, color=colors[i], label=skill)
        ax.set_xlabel('Percentage')
        ax.set_title(f'Skill: {skill}')
        ax.legend(loc='best')

    # Hide any empty subplots if the number of skills is not a multiple of num_columns
    for i in range(num_skills, num_rows * num_columns):
        row = i // num_columns
        col = i % num_columns
        fig.delaxes(axes[row, col])

    plt.tight_layout()
    plt.subplots_adjust(top=0.9)
    plt.show()
```

```python
def multilabel_train_test_split(*arrays,
                                test_size=None,
                                train_size=None,
                                random_state=None,
                                shuffle=True,
                                stratify=None):
    """
    Train test split for multilabel classification. Uses the algorithm from: 
    'Sechidis K., Tsoumakas G., Vlahavas I. (2011) On the Stratification of Multi-Label Data'.
    """
    if stratify is None:
        return train_test_split(*arrays, test_size=test_size,train_size=train_size,
                                random_state=random_state, stratify=None, shuffle=shuffle)
    
    assert shuffle, "Stratified train/test split is not implemented for shuffle=False"
    
    n_arrays = len(arrays)
    arrays = indexable(*arrays)
    n_samples = _num_samples(arrays[0])
    n_train, n_test = _validate_shuffle_split(
        n_samples, test_size, train_size, default_test_size=0.25
    )
    cv = MultilabelStratifiedShuffleSplit(test_size=n_test, train_size=n_train, random_state=123)
    train, test = next(cv.split(X=arrays[0], y=stratify))

    return list(
        chain.from_iterable(
            (_safe_indexing(a, train), _safe_indexing(a, test)) for a in arrays
        )
    )
```

```python
X_train_1, X_test_1, y_train_1, y_test_1 = multilabel_train_test_split(X_1, y_1, stratify=y_1, test_size = 0.3)
```

```python
scaler = MinMaxScaler()
classifier = MultiOutputClassifier(KNeighborsClassifier())
cv = RepeatedKFold(n_splits=5, n_repeats=5, random_state=42)
```

```python
pipeline = Pipeline([
    ('scaler', scaler),           
    ('classifier', classifier)  
])

# Step 4: Define the hyperparameter grid for tuning
param_grid = {
    'classifier__estimator__n_neighbors': [5, 7, 9],
    'classifier__estimator__weights': ['uniform', 'distance']
}


scoring = make_scorer(hamming_loss, greater_is_better=False)

grid_search = GridSearchCV(pipeline, param_grid, cv=cv, scoring=scoring)

grid_search.fit(X_train_1, y_train_1) 


best_pipeline = grid_search.best_estimator_
best_params = grid_search.best_params_
```
