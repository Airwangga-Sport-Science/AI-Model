```python
import pandas as pd 

df = pd.read_csv(r'D:\Program\Airwangga Sport Science\AI Model\AI-Model\Data\FM 2023.csv') #Sesuaikan Path dengan lokasi Dataset di masing-masing device
```

```python
df.set_index(df['Name'].values, inplace=True)
```

```python
df
```

```python
X = df[['Corners', 'Crossing', 'Dribbling', 'Finishing', 'First Touch', 'Free Kick Taking', 'Heading', 'Long Shots',
   'Long Throws', 'Passing', 'Penalty Taking', 'Tackling', 'Acceleration', 'Agility',
    'Jumping Reach', 'Natural Fitness', 'Pace', 'Stamina', 'Strength']]
y = df[['GK', 'DL', 'DC', 'DR', 'WBL', 'WBR', 'DM', 'ML', 'MC', 'AML', 'AMC', 'AMR', 'ST']]
```

```python
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
vectors = X.values
scaled_rows = scaler.fit_transform(vectors.T).T
X_normalized = pd.DataFrame(data = scaled_rows, columns = X.columns)
```

```python
X_normalized.set_index(df['Name'].values, inplace=True)
```

```python
X_normalized
```

```python
#Non Neural Network Regression
import matplotlib
from matplotlib.collections import PatchCollection
from matplotlib.patches import Circle
from matplotlib.colors import ListedColormap
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor

Knn=KNeighborsRegressor()                          # Accept default parameters
LR=LinearRegression()
RF=RandomForestRegressor(max_depth=None, max_features=None, random_state=42)


Knn.fit(X_normalized, y)   
LR.fit(X_normalized, y) 
RF.fit(X_normalized, y) 
```

```python
#Neural Network Regression

from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, InputLayer, Flatten
from tensorflow.keras.callbacks import EarlyStopping

NNetModel = Sequential()

NNetModel.add(Dense(len(X_normalized.columns), activation='relu', input_dim=len(X_normalized.columns)))
NNetModel.add(BatchNormalization())
NNetModel.add(Dense(16, activation='relu'))
NNetModel.add(Dropout(0.2))
NNetModel.add(Dense(32, activation='relu'))
NNetModel.add(Dropout(0.5))
NNetModel.add(Dense(len(y.columns), activation='linear'))

NNetModel.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse'])
es = EarlyStopping(monitor='loss', mode='min', verbose=1, patience=10)
NNetModel.fit(X_normalized.values, y.values, batch_size=64, epochs=4000, verbose=1, validation_split=0.3, callbacks=[es])
```

```python
def predict_player_position(player_name, model):

    positions = model.predict(X_normalized.loc[[player_name]].values)

    R_scaler = preprocessing.MinMaxScaler(feature_range=(1, 20))
    vectors = pd.DataFrame(positions.reshape(1,13), columns=y.columns).values
    scaled_rows = R_scaler.fit_transform(vectors.T).T

    results = pd.DataFrame({'Position':y.columns, 'Real':y.loc[player_name], 'Predicted':scaled_rows[0], 'Difference':y.loc[player_name]-scaled_rows[0]})

    return results.nlargest(n=3, columns=['Predicted'])
```

```python
import numpy as np
def predict_and_show(player_name, model):
    results = predict_player_position(player_name, model)
    print(results)
    predicted = model.predict(X_normalized.values)
    mse = np.mean((y.values - predicted) ** 2)
    print("\nMSE:", mse)
```
